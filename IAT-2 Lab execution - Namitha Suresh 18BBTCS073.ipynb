```python
#i. Use Data Wrangling to load this dataset
import pandas as pd
from matplotlib import pyplot as plt 
df1=pd.read_csv('NationalNames.csv')
df1
```

```python
df2=pd.read_csv('StateNames.csv')
df2
```

```python
#ii. Find sum of the birth’s column by sex as the total number of births in that year
total_births = df1.pivot_table('Count', index='Year',
                                columns='Gender', aggfunc='sum')
total_births
```

```python
#iv.Visualize total births by sex and year
graph = total_births.plot(title='Total births categorised by sex and year')
print(graph)
```
