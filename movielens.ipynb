```python
#i. Load each table into a pandas data frame object using pandas.read_table
import pandas as pd
import numpy as np

df1=pd.read_csv("Movies.csv",sep="::",encoding='latin-1',engine='python',names=['MovieID','Title','Genres'])
df2=pd.read_csv("ratings.csv",sep="::",encoding='latin-1',engine='python',names=['UserID','MovieID','Rating','Timestamp'])
df3=pd.read_csv("users.csv",sep="::",encoding='latin-1',engine='python',names=['UserID','Gender','Age','Occupation','Zip-code'])
print(df1.head(),'\n\n',df2.head(),'\n\n',df3.head(),'\n\n')

#outer join - Merging ratings with users
outer_merged = pd.merge(df2,df3,how="outer")
outer_merged.head()
```

```python
#ii. Merge ratings with users and then merge that result with the movies data
final_merged = pd.concat([df1,outer_merged],axis=1)
final_merged.head()
```

```python
#iii. Calculate mean movie ratings for each film grouped by gender
df=final_merged
grouped = df.groupby('Gender')
print(grouped['Rating'].agg(np.mean))
```
